```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,100)
y = np.linspace(-5,5,100)

X,Y = np.meshgrid(x,y)
```

```python
plt.scatter(x,y,s=1)
```

```python
r = np.sqrt(X**2 + Y**2) # r=0 at origin
V = 1/r #Potential due to point charge
```

```python
r[r==0] = 0.01
```

```python
plt.contourf(X,Y,V, 100)
plt.colorbar()
```

```python
#center becomes bright/high color which means high potential 
#Whereas as it moves further the regions become darker/lower in color meaning less potential
```

```python
Ex = X/(r**3)

Ey = Y/(r**3)
```

```python
plt.quiver(X,Y,Ex,Ey)
charges = [(+1,-1,0), (-1,1,0)]
plt.contour(X,Y,V)
plt.scatter(0,0,color='red',s=100)
```
