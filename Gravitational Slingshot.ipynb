```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Gravity constants
G = 1
M = 500

# Time step
dt = 0.01

# Initial position
x = -10
y = 3

# Initial velocity
vx = 0
vy = 2



# Lists to store trajectory
x_list = []
y_list = []

# Simulation loop
for i in range(5000):

    # Distance from planet
    r = np.sqrt(x**2 + y**2)

    # Gravitational acceleration
    ax = -G*M*x/(r**3)
    ay = -G*M*y/(r**3)

    # Update velocity
    vx += ax*dt
    vy += ay*dt

    # Update position
    x += vx*dt
    y += vy*dt

    # Store points
    x_list.append(x)
    y_list.append(y)

# Plot trajectory
plt.figure(figsize=(7,7))

plt.plot(x_list, y_list, color='cyan')

# Planet
plt.scatter(0,0,color='orange',s=300)

plt.xlabel("x")
plt.ylabel("y")

plt.title("Gravitational Slingshot Simulation")

plt.grid()

plt.axis('equal')

plt.show()
```
